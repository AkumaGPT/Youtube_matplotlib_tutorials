COMMON PLOT TYPES (LINE PLOT, SCATTER PLOT, BAR CHART, HISTOGRAM)

IMPORTING LIBRARY

```python
import matplotlib.pyplot as plt 
```

DEMO 1 - LINE AND SCATTER PLOT

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 3))

# line plot: checking your battery level over the hours
hours = [1, 2, 3, 4, 5]
battery = [95, 80, 65, 50, 35] # percentage
ax[0].plot(hours, battery)

# scatter plot: check the relationship between study hours and test score
study_hours = [2, 3, 1, 4, 5]
test_score = [60, 70, 59, 80, 20]
ax[1].scatter(study_hours, test_score)

plt.tight_layout()
plt.show()
```

DEMO 2 - BAR CHART AND HISTOGRAM

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 3))
# Bar chart: check your monthly spending by category
categories = ["food", "transport", "others"]
spending = [20000, 12000, 8000]
ax[0].bar(categories, spending)


# Histogram: check your daily commute time
commute_time = [10, 12, 12, 15, 18, 20, 20, 25, 30, 35, 35, 40] # in minutes

# count = 40 - 10
# step = int(count / 6)
# bins = [20, 25, 30, 35, 40]

ax[1].hist(commute_time, bins=5, edgecolor="w")
ax[1].axvline(25, color='r', linestyle='dashed')

plt.tight_layout()
plt.show()

# math
# count = 40 - 10 = 30
# step = 30 / 5 = 6
# bins = [10, 16, 22, 28, 34, 40]


```

SOLO CHALLENGE

```python
import numpy as np

fig, ax = plt.subplots(1, 2, figsize=(12, 3))
estimated = [10, 12, 9, 15, 14, 13] # your budget for the day
actual    = [9, 11, 10, 14, 13, 13] # what you actually spent

# left plot: scatter plot
ax[0].scatter(estimated, actual)
ax[0].plot([9, 15], [9, 15])

# right plot: histogram of errors (Actual - Estimated)
error = (np.array(actual) - np.array(estimated)).tolist()
ax[1].hist(error, bins=3, edgecolor='white')
ax[1].axvline(0, color='r', linestyle=':')

plt.tight_layout()
plt.show()
```
