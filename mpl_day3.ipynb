IMPORTING LIBRARY

```python
import matplotlib.pyplot as plt
```

DEMO 1 - LINE AND SCATTER PLOT

```python
# line plot: checking your battery level over the hours
hours = [1, 2, 3, 4, 5]
battery = [95, 80, 65, 50, 35] # percentage
fig, ax = plt.subplots(1, 2, figsize=(12, 3))
ax[0].plot(hours, battery)

# scatter plot: check the relationship between study hours and test score
study_hours = [2, 3, 1, 4, 5]
test_score = [60, 70, 59, 80, 20]
ax[1].scatter(study_hours, test_score)

plt.tight_layout()
plt.show()
```

DEMO 2 - BAR CHART AND HISTOGRAM

```python
# Bar chart: check your monthly spending by category
categories = ["food", "transport", "others"]
spending = [20000, 12000, 8000]
fig, ax = plt.subplots(1, 2, figsize=(12, 3))
ax[0].bar(categories, spending)

# Histogram: check your daily commut time
commute_time = [10, 12, 12, 15, 18, 20, 20, 25, 30, 35, 35, 40] # in minutes
count = 40 - 10
step = int(count / 5)
bins = list(range(10, 41, step))

ax[1].hist(commute_time, bins=bins, edgecolor='w')

plt.tight_layout()
plt.show()

# math
# count = 40 - 10 = 30
# step = 30 / 5 = 6
# bins = [10, 16, 22, 28, 34, 40]

# code
# count = 40 - 10
# step = int(count / 5)
# bins = list(range(10, 41, step))
# bins
```
