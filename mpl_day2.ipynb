FIGURE ANATOMY AND SUBPLOTS

IMPORTING LIBRARY

```python
import matplotlib.pyplot as plt
```

DEMO 1

```python
fig, ax = plt.subplots(1, 2)

#left subplot: check your savings
ax[0].plot([1, 2, 3], [1000, 4000, 9000])

#right subplot: check your water intake
ax[1].plot([1, 2, 3], [1, 2, 3])

# show the image
plt.show()
```

DEMO 2

```python
fig, ax = plt.subplots(2, 2)

# top left subplot: check study hours remaining
ax[0, 0].plot([1, 2, 3], [1, 4, 9])

# top right: check your remaining data
ax[0, 1].plot([1, 2, 3], [3, 2, 1])

# bottom left subplot: check your transport fare
ax[1, 0].plot([1, 2, 3], [2, 2, 2])

# bottom right subplot: check social media followers growth
ax[1, 1].plot([1, 2, 3], [1, 8, 27])

# show the figure
plt.show()
```

SOLO CHALLENGE

```python
days = [1, 2, 3, 4, 5]
morning_temp = [26, 25, 25, 24, 24]
afternoon_temp = [31, 32, 33, 34, 33]

# create a 1x2 grid
fig, ax = plt.subplots(1, 2)

# left subplot: days vs morning_temp
ax[0].plot(days, morning_temp)

# right subplot: days vs evening_temp
ax[1].plot(days, afternoon_temp)


# show the figure
plt.show()
```
